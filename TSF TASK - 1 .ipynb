```python
# IMPORTING REQUIRED LIBRARIES #

import numpy as np
import  pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("http://bit.ly/w-data")
print("Data: ")
dataset.head()
```

```python
# VISUALIZING DATA TO ESTABLISH RELATION IN OUR BETWEEN DATA #

dataset.plot(x="Hours", y="Scores" , style="o")
plt.title("Hours V/S Score")
plt.xlabel("Hours of Study")
plt.ylabel("Marks Scored")
plt.show
```

```python
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,1].values
print(x)
print(y)
```

```python
# SPLITTING DATA SET USING SK-LEARN #

from sklearn.model_selection import train_test_split
x_train, x_test , y_train, y_test = train_test_split(x,y, train_size=0.80, test_size=0.20, random_state=0)
```

```python
# TRAINING THE MODEL #

from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train , y_train)
print("THE MODEL IS READY")
```

```python
plt.scatter(x_train, y_train , color="Red")
plt.plot(x_train, lr.predict(x_train),color="Black")
plt.title("Hours V/S Score")
plt.xlabel("Hours of Study")
plt.ylabel("Marks Scored")
plt.show
```

```python
plt.scatter(x_test, y_test , color="Red")
plt.plot(x_train, lr.predict(x_train),color="Yellow")
plt.title("Hours V/S Score")
plt.xlabel("Hours of Study")
plt.ylabel("Marks Scored")
plt.show
```

```python
# PREDICTION USING TRAINED MODEL #

output=lr.predict([[9.25]])
print(output)
```

```python
# EFFICIENCY OF OUR MODEL #

score=lr.score(x_train, y_train)
print(score*100, "%")
```
